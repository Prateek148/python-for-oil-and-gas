```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import math as m
import plotly.express as pp
```

```python
t = np.arange(0,1200,200)
D = .005
```

```python
#Exponential
Qt_1 = 200*(2.718**(-D*t))
Qt_1
```

```python
#hyperbolic
Qt_2 = 200/((1+(D*t*.5))**(2))
Qt_2
```

```python
#Harmonic
Qt_3 = 200/(1+(.005*t))
Qt_3
```

```python
plt.style.use('default')
plt.figure(figsize=(10,5))
plt.plot(t,Qt_1,label='Exponential')
plt.plot(t,Qt_2,label='Hyperbolic')
plt.plot(t,Qt_3,label='Harmonic')
plt.xlabel('Time(Days)')
plt.ylabel('Flow Rate (bbl/day)')
plt.title('Production Decline curve')
plt.legend()
plt.grid() 
```
