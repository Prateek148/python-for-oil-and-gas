```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import math as m
```

```python
kro = np.array((.8,.610,.470,.370,.285,.220,.163,.120,.081,.050,.027,.010,0))
krw = np.array((0,.002,.009,.020,.033,.051,.075,.10,.132,.170,.208,.251,.3))
Sw = np.array((.2,.25,.3,.35,.4,.45,.5,.55,.6,.65,.7,.75,.8))
#Sw = np.linspace(.001,1,1000)
```

```python
#Case1
#R1 = np.array((0,.247,.657,.844,.921,.959,.979,.988,.994,.997,.999,.999,1))
#R1 = uo/uw for case 1 , similarly for R2 & R3 for case 2 and case 3
#Case2
#R2 = np.array((0,.032,.161,.351,.537,.699,.821,.893,.942,.971,.987,.996,1))
#Case3
#R3 = np.array((0,.001,.008,.021,.044,.085,.155,.250,.394,.576,.755,.909,1))
```

```python
R1 = .01 #uo/uw
R2 = .1 #uo/uw
R3 = 2.4 #uo/uw
```

```python
#Case 1
Fw1 = 1/(1+(kro*R1)/krw)
Fw1
```

```python
#Case 2
Fw2 = 1/(1+(kro*R2)/krw)
Fw2
```

```python
#Case 3
Fw3 = 1/(1+(kro*R3)/krw)
Fw3
```

```python
plt.style.use('default')
plt.figure(figsize=(10,5))
plt.xlim(0,.9,10)
plt.xlabel('Water Saturation')
plt.ylim(0,1,.2)
plt.ylabel('Fractional Flow (Fw)')
plt.plot(Sw,Fw1,label='uo/uw = .01',linewidth = '4')
plt.plot(Sw,Fw2,label='uo/uw=.1',linewidth = '4')
plt.plot(Sw,Fw3,label='uo/uw=2.5',linewidth ='4')
plt.grid()
plt.title('Fw vs Saturation')
plt.legend()
```
