```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Sw = np.linspace(0,1,100)
Krw = Sw**3 ; Kro = (1-Sw)**3
plt.style.use('default')
plt.plot(Sw,Krw,lw =4,color='red')
plt.plot(Sw,Kro,lw=4,color='blue')
plt.grid()
plt.title('Saturation vs Relative Permeability')
plt.xlabel('Saturation')
plt.ylabel('Relative Permeability')
plt.legend()

```

```python
P = pd.read_excel('/content/data.1.xls')
P = P.replace('-999.25',np.nan)
P
```

```python
P.describe()
```

```python
plt.style.use('default')
plt.figure(figsize=(4,8))
plt.plot(P['GR'],P['DEPT'])
plt.plot(P['SP'],P['DEPT'])
plt.ylim(3400,100)
plt.xlim(0,600)
plt.axvline(x=200,color='red')
```

```python
P = P.replace('-999.25',np.nan)
plt.style.use('default')
plt.figure(figsize=(4,15))
plt.plot(P['NPLS'],P['DEPT'],label='Neutron Porosity')
plt.plot(P['DPLS'],P['DEPT'],label='Lithology Density Porosity')
plt.ylim(3400,200)
#plt.axvline(x=20,color='red')
plt.xlim(0,80)
plt.legend()
```

```python
P['Vsh']=(P.GR - P.GR.min())/(P.GR.max() - P.GR.min())
P['Eff. Porosity']=(1-P['Vsh'])
fig, R = plt.subplots(nrows=1,ncols=4)
fig.set_figheight(15)
fig.set_figwidth(15)
P=P.replace('-999.25',np.nan)
#R[0].plot(P['CALM'],P['DEPT'],label='Caliper (Diameter)')
R[0].plot(P['GR'],P['DEPT'],label='Gamma Ray',color='Green')
R[0].plot(P['SP'],P['DEPT'],label='Self spontaneous',color='Pink')
R[0].set_ylim(3380,200)
R[0].set_xlim(-200,500)
R[0].set_title('Lithology log')
R[0].legend(loc='Upper right')
R[0].grid()

#R[1].twiny()
R[1].plot(np.log(P['ILD']),P['DEPT'],label='Deep Induction')
R[1].plot(np.log(P['ILM']),P['DEPT'],label='Medium Induction')
R[1].set_ylim(3380,200)
R[1].set_xlim(0,10)
R[1].set_title('Resistivity log (ohmm)')
R[1].grid()
R[1].legend()

R[2].plot(P['PE'],P['DEPT'])
R[2].set_title('PE log (b/e-)')
R[2].set_ylim(3380,200)
R[2].set_xlim(0,10)
R[2].grid()
R[2].legend()

#R[3].plot(P['NPLS'],P['DEPT'],label='Compensated Neutron Porosity',color='Black',alpha=.6)
#R[3].plot(P['DPLS'],P['DEPT'],label='Lithology Density Porosity',color='Blue',alpha=.6)
#R[3].twiny()
R[3].set_xlim(0,80)
R[3].plot(P['NPLS'],P['DEPT'],label='NPLS',color='red')
R[3].plot(P['DPLS'],P['DEPT'],label='DPLS',color='blue',alpha=.3)
R[3].set_ylim(3380,200)
R[3].set_title('Porosity log')
R[3].grid()
R[3].legend(loc='Upper right')

R[4].twiny()
R[4].plot(P['Vsh'],P['DEPT'])
R[4].set_ylim(3380,200)
R[4].set_xlim(0,1)
R[4].set_title('Volume of shale')
R[4].grid()
R[4].legend()

R[5].plot(P['Eff. Porosity'],P['DEPT'],label='Eff Porosity',color='gold')
R[5].set_ylim(3380,200)
R[5].set_xlim(0,1)
R[5].set_title('Effective Porosity')
R[5].grid()
R[5].legend()

```
